```python
#TFIDF
import nltk 
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

stop_words= set(stopwords.words('english'))
ps= PorterStemmer()

corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning is fun and exciting.",
    "Natural language processing helps computers understand text.",
    "ChatGPT is an AI developed by OpenAI.",
    "The weather today is sunny with a slight chance of rain.",
    "I love pizza, pasta, and all kinds of Italian food.",
    "Python is a powerful programming language for data science.",
    "Books and libraries are sources of knowledge and wisdom.",
    "Social media has changed how people communicate.",
    "Electric cars are better for the environment than gasoline cars."
]
st=[]
for doc in corpus:
    tk = word_tokenize(doc)
    sp=[]
    for word in tk:
        if word not in stop_words:
            stem= ps.stem(word)
            sp.append(stem)
    st.append(" ".join(sp))
print(st)
tdidf= TfidfVectorizer().fit_transform(st)
print(tdidf)
            
```
